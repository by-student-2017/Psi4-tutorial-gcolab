```python
# import Psi4
import psi4
# check Psi4 version
print(psi4.__version__)
```

```python
#! 6-31G** H2O Test RASSCF Energy Point
#! will default to only singles and doubles in the active space
psi4.core.set_output_file("log.txt")

psi4.set_memory('10GB')
psi4.set_num_threads(2)

mol = psi4.geometry("""
    O
    H 1 1.00
    H 1 1.00 2 103.1
""")

psi4.set_options({
    'restricted_docc': [1, 0, 0, 0],
    'ras1': [2, 0, 1, 1],
    'ras3': [1, 0, 0, 1],
    'restricted_uocc': [8, 2, 3, 5],
    'mcscf_maxiter': 80,
    'mcscf_r_convergence': 5.e-6
})

scf_e, scf_wfn = psi4.energy("SCF/6-31G**", return_wfn=True)
```

```python
psi4.set_options({
    'icore': 1
})
rasscf_energy, rasscf_wfn = psi4.energy('RASSCF/6-31G**', ref_wfn=scf_wfn, return_wfn=True)
```

```python
psi4.set_options({
    'icore': 0
})
rasscf_energy = psi4.energy('RASSCF/6-31G**', ref_wfn=rasscf_wfn)
```
